```python
import warnings
warnings.filterwarnings("ignore", category=DeprecationWarning)
import pandas as pd
import numpy as np
from matplotlib import cm, pyplot as plt
from matplotlib.dates import YearLocator, MonthLocator
from hmmlearn.hmm import GMMHMM

btc = pd.read_csv('./data/bitcoin_price.csv')

btc.head()
```

```python
btc = btc.iloc[::-1]

btc.head()
```

```python
features = btc[["Open","High","Low","Close"]].values

features[:4]
```

```python
fracChange = ((features[:,3]/features[:,0]) - 1)*100
fracHigh   = ((features[:,1]/features[:,0]) - 1)*100
fracLow    = (1 - (features[:,2]/features[:,0]))*100

X = np.column_stack([fracChange, fracHigh, fracLow])

train = X[:1647]
train[:4]
```

```python
model = GMMHMM(n_components=4, n_mix=6, n_iter=400, tol=0.001, verbose=True)
model.fit(train)
```

```python
predFracChange = np.linspace(-0.1, 0.1, num=50)
predFracHigh   = np.linspace(0, 0.1, num=10)
predFracLow    = np.linspace(0, 0.1, num=10)

observations = []
for i in predFracChange:
    for j in predFracHigh:
        for k in predFracLow:
            observations.append([i, j, k])

observations = np.array(observations)
observations[:4]
```

```python
observationSeq = train[-10:]
observationSeq
```

```python
score = -9999999
best = []
for i in observations:
    seq = observationSeq
    np.vstack([seq, i])
    s = model.score(seq)
    #print(s)
    if(s >= score):
        best = i
        score = s
        print(best, score)
```
